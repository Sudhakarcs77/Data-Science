## **16 Join**

```python
A = pd.DataFrame({'key':[1,2,4,5,6],
                  'firstName' : ['ram','syam','bhim','ravan','laxman'],
                  'lastName': ['x','y','z','a','b']
}
                 )

```

```python
B = pd.DataFrame({'key':[1,2,3,7,8],
                  'firstName' : ['Rana','Prabhas','AA','NTR','charan'],
                  'lastName': ['m','l','n','o','p']
}
                 )
```

```python
A.merge(B,on='key')
```

```python
A.merge(B,on='key',how='outer')
```
